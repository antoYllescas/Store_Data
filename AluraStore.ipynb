#**Análisis gráfico**

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_t1, ingreso_t2, ingreso_t3, ingreso_t4]
```

##**Ingresos**

```python
fondo_app = '#473472'
color_efi = '#AD49E1'
color_def = '#EB3678'

peor_ingreso = min(ingresos)
colores_barras = [color_def if ingresos == peor_ingreso else color_efi for ingresos in \
                  ingresos]
fig, ax = plt.subplots(figsize=(8, 5))
fig.patch.set_facecolor(fondo_app)
ax.set_facecolor(fondo_app)

barras = ax.bar(tiendas, ingresos, color=colores_barras, edgecolor='white', linewidth=1)

ax.bar_label(barras, fmt='${:,.0f}', label_type='center', color='white', fontweight='bold', fontsize=9)

ax.tick_params(colors='white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.set_title('Reporte de Eficiencia: Ingresos por Tienda', color='white', pad=20, fontsize=14, fontweight='bold')
ax.set_ylabel('Ingreso en $ (Billones)', color='white')

plt.show()
```

**Resumen Ejecutivo**

Tras un análisis exhaustivo de las métricas de ingresos, satisfacción del cliente, rotación de inventario y participación de mercado a nivel nacional de las 4 sucursales, la recomendación fundamentada en datos es proceder con la venta de la Tienda 4.

Durante la auditoría descubrimos una anomalía comercial clave: la satisfacción del cliente no es el motor principal de los ingresos en el ecosistema actual de Alura Store. Por lo tanto, la decisión de liquidación se basó en la ineficiencia financiera y la pérdida de territorio, no en las calificaciones de servicio.

**Hallazgos Clave que Sustentan la Decisión:**
1. Desempeño de Ingresos (El Factor Crítico)
La Tienda 4 representa el eslabón más débil de la cadena a nivel de facturación, generando el ingreso más bajo del grupo con $1,038,375,700

 En contraste, la Tienda 1 lidera la captación de capital superando los $1.15 billones.

2. El Falso Positivo de la Satisfacción
Inicialmente, la Tienda 1 presentaba la calificación más baja de la cadena (3.97 estrellas). Sin embargo, cruzando esta variable con sus ingresos, concluimos que cuenta con un mercado cautivo: los clientes siguen comprando a pesar del mal servicio. Por otro lado, la Tienda 4 mantiene una calificación estándar aceptable (casi 4 estrellas), pero esta "buena atención" no se está traduciendo en ventas, lo que indica un problema estructural más profundo.

3. Inventario Estancado (Capital Muerto)
